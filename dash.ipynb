```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/clean/df_total.csv')
df['DATE'] = pd.to_datetime(df['DATE'])
df = df.set_index('DATE')
df = df.resample('YE').mean()
df = df[['T_Q']]
df = df.rename(columns={'T_Q':'Temp [m]'})
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Temp [m]'], linewidth=2.5, alpha=0.85)  # smooth line

# Minimalism: remove right/top spines, keep left/bottom only
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_linewidth(1)
ax.spines['bottom'].set_linewidth(1)

# Soft grid, no axis background
plt.grid(axis='y', color='#CCCCCC', linestyle=':', linewidth=1, alpha=0.4)
plt.gcf().set_facecolor('white')
ax.set_facecolor('white')

# Titles, labels
plt.title('Temperature Over Time', fontsize=18, weight='bold', pad=20)
plt.ylabel('Temp [C]', fontsize=14, labelpad=15)
plt.xlabel('Date', fontsize=14, labelpad=10)

# Ticks styling
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout(pad=2)

# Save or show
plt.show()
```
